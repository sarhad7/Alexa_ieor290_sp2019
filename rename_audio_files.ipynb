```python
# Import Packages
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Read the metadata into a Dataframe 
csv_filename = "vox1_meta.csv"
df = pd.read_csv(csv_filename, delimiter = "\t")

# Split the info into ID, Author Name, Gender (m,f), Nationality
wav_ids = []
authors = []
genders = []
nationalities = []
for label in df.values:
    wav_ids.append(label[0].split("\t")[0])
    authors.append(" ".join(label[0].split("\t")[1].split("_")))
    genders.append(label[0].split("\t")[2].upper())
    nationalities.append(label[0].split("\t")[3])
    
# Put the info in an organized dataFrame df_info
df_info = pd.DataFrame()
df_info["ID"] = wav_ids
df_info["Name"] = authors
df_info["Gender"] = genders
df_info["Nationality"] = nationalities
df_info.set_index("ID", inplace = True)
```

```python
# Every .wav file of the test utterances downloaded from the voxceleb website is contained in a randomly named folder 
#  (i.e 5r0dWxy17C8). Each of these randomly named folders contain 0 or more utterances and are inside a parent folder 
# named by the id of the speaker (i.e id10309).The following function searches inside the directory that contains all id##### 
#  folders and renames the audio files sequentially. You can comment out the "print" lines - They are included for illustration
#  In my laptop, the parent directory that contains all id##### folders is as it appears below

directory = "D:/UC Berkeley/Graduate/Classes/Spring 2019/IEOR290/Alexa/rename_audio_files/wav" # Specify Parent Directory Here
mid_syllab = "_split_sent"

def rename_audio_files(directory):
    for dirname in os.listdir(directory): # parse through file list in the current directory
        n = 1
        print("checking inside {}".format(dirname))
        for subdirname in os.listdir(directory+"/"+dirname): # parse through file list in the sub directory
            print("checking inside {}".format(subdirname))
            for filename in os.listdir(directory+"/"+dirname+"/"+subdirname):
                name = df_info.loc[dirname].Name.split()
                init_syllab = name[0][0].lower()+name[-1].lower()
                new_filename = init_syllab+mid_syllab+"{0:0=3d}".format(n)+".wav"
                print("Renaming {} to {}".format(filename, new_filename))
                os.rename(directory+"/"+dirname+"/"+subdirname+"/"+filename, new_filename)
                n += 1
```

```python
rename_audio_files(directory) # run the conversion
```
